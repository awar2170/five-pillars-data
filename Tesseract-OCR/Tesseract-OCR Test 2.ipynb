```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load image
image_path = "/mnt/data/5e05ed2d-abc7-4b06-a938-6024d7879da8.jpg"
img = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Threshold to binary
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

# Find contours
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Loop through contours and identify square-like boxes
for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    aspect_ratio = w / float(h)
    area = cv2.contourArea(cnt)

    if 10 < w < 30 and 10 < h < 30 and 0.8 < aspect_ratio < 1.2:
        roi = thresh[y:y+h, x:x+w]
        non_zero_pixels = cv2.countNonZero(roi)
        fill_ratio = non_zero_pixels / (w * h)

        status = "Checked" if fill_ratio > 0.3 else "Empty"
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(img, status, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (36,255,12), 1)

# Show result
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Checkbox Detection")
plt.axis("off")
plt.show()
```
