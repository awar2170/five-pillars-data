This is the structure I need for the item properties. If a property has an upcast effect, it needs to be duplicated with the appropriate field changes and the upcast bool flipped.

The only destroyable properties are Enchantments, Curses, and Blessings. Value is generally a number, and is normally 1/2/3/4. 


Notes: 
- Property_size = capacity (if ORACLE, then capcity = 1; unless Boon, then capcity is 0)
- capacity for mystic and taskmaster -- all 0

property uniqueness (if 1 = True, 0 = False)
Art: All are true 
Orc: All are true
Enchanter: Enchat are all false, curses are all true 
Task/Myst: all false 

````python
# Pull in the df
df = pd.read_excel('input/--DEV-- Tablet Library.xlsx', sheet_name="ARTIFICER", skiprows = 1)
# df
````

````python
df
````

````python
# Create a simple JSON structure
# Note: will need to add iterations as needed (ex: pillar_names[x])
One_JSON_Object = { 
                  "property_name": df["Name"][0],
                  "property_size": True if df["property_size"][0] == 1 else False, #All art tabs have cap 0 (even if it's the cap. effecting tablets)
                  "property_source": "Tablet",
                  "property_ability": None, #keep null
                  "property_monster": None, #keep null
                  "property_class": "Artificer", #class name: artificer, oralce, enchanter, etc.
                  "property_computation": df["Computational Name"][0], #computational name
                  "property_value": df["Property Value"][0] if df["Property Value"][0] != None else None, #maps to property_value in excel
                  "property_description": df["Description"][0],
                  "property_limitations": df["Limitations"][0], #maps to limitations
                  "boon_effect": df["Effect"][0] if df["Computational Name"][0] == "Boon" else None,
                  "property_unique": True if df["property_unique"][0] == 1 else False, #See notes 
                  "property_curse": True if df["Computational Name"][0] == "Curse" else False, #is it a curse? Yes, true; no, false
                  "property_destroyable": True if df["property_destroyable"][0] == 0 else False, #blessings, curses, enchantments are destroyable (are true) everything else is false
                  "property_hidden": True if df["Computational Name"][0] == "Curse" else False, #all curses true, everything else false
                  "upcast": False #change later
}
json.dumps(One_JSON_Object, indent=2)
    
````

````python
print(json.dumps(One_JSON_Object, indent=2))
````

````python
pillar_names = ["ARTIFICER", "ENCHANTER", "ORACLE", "TASKMASTER", "MYSTIC"]

for x in np.arange(0,len(pillar_names),1):
    df = pd.read_excel('input/--DEV-- Tablet Library.xlsx', sheet_name=f'{pillar_names[x]}', skiprows = 1)
    tab = pillar_names[x]
    tablet_JSON = []
 
    for y in np.arange(0, len(df),1):
        One_JSON_Object = { 
                          "property_name": df["Name"][y],
                          "property_size": True if df["property_size"][y] == 1 else False, #All art tabs have cap 0 (even if it's the cap. effecting tablets)
                          "property_source": "Tablet",
                          "property_ability": None, #keep null
                          "property_monster": None, #keep null
                          "property_class": pillar_names[x], #class name: artificer, oralce, enchanter, etc.
                          "property_computation": df["Computational Name"][y] if "Computational Name" in df.columns else None, #computational name
                          "property_value": df["Property Value"][y] if "Property Value" in df.columns else None, #maps to property_value in excel
                          "property_description": df["Description"][y],
                          "property_limitations": df["Limitations"][y], #maps to limitations
                          "boon_effect": df["Effect"][y] if "Computational Name" in df.columns and df["Computational Name"][y] == "Boon" else None,
                          "property_unique": True if df["property_unique"][y] == 1 else False, #See notes 
                          "property_curse": True if "Computational Name" in df.columns and df["Computational Name"][y] == "Curse" else False, #is it a curse? Yes, true; no, false
                          "property_destroyable": True if df["property_destroyable"][y] == 0 else False, #blessings, curses, enchantments are destroyable (are true) everything else is false
                          "property_hidden": True if "Computational Name" in df.columns and df["Computational Name"][y] == "Curse" else False, #all curses true, everything else false
                          "upcast": False #change later
        }        
        tablet_JSON.append(One_JSON_Object)
        
    file_name = tab + "_tablet.json"
    
    # print(f"{file_name} successfully created")
    # print(tablet_JSON)

    with open(file_name, "w") as fp:
        json.dump(tablet_JSON, fp, indent=2, ensure_ascii = True, allow_nan = True)
    print(f"{file_name} created")
    
````

## Put it All Together!

````python
# Tabs of the Excel Sheet
tab_names = ["ARTIFICER", "ENCHANTER", "ORACLE"]
null = None

for x in np.arange(0, len(tab_names),1):
    df = pd.read_excel('--DEV-- Tablet Library.xlsx', sheet_name=f'{tab_names[x]}', skiprows = 1)
    tab = tab_names[x]
    hidden_boolean = False
    tablet_JSON = []
    
    # Collect Row Information 
    for y in np.arange(0, len(df),1):
        item_name = str(df["Name"][y])
        computation_name = str(df["Computational Name"][y])
        description_text = str(df["Description"][y])
        unique_str = str(df["Uniqueness"][y])
        computation_limits = str(df["Computational Limits"][y])
        limitations = str(df["Limitations"][y])
        grade = int(df["Grade"][y])
                
        if tab == "ENCHANTER" or tab == "ORACLE":
            destroy_boolean = True
        else: 
            destroy_boolean = False

        
        # Curse
        if "Curse" in item_name:
            curse_boolean = True
        else:
            curse_boolean = False
        
        
        # Capcity Effect
        if "Capacity Effect" in df:
            capacity = int(df["Capacity Effect"][y])*-1
        else: 
            capacity = None
    
    # Create the JSON Object 
        One_JSON_Object = {"property_name": item_name, "property_computation_name": computation_name, 
                           "property_grade": grade, "property_size": capacity, "property_source": tab, 
                           "property_ability": null, "property_monster": null, "property_type": null, 
                           "property_computation": null, "property_value": null, 
                           "property_limitations": limitations, "property_computation_limits": computation_limits, 
                           "property_description": description_text, "property_unique": unique_str,
                           "property_curse": curse_boolean, "property_destroyable": destroy_boolean, 
                           "property_hidden": hidden_boolean}

        tablet_JSON.append(One_JSON_Object)
        
    file_name = tab + "_tablet.json"
    
    # print(f"{file_name} successfully created")
    # print(tablet_JSON)

    with open(file_name, "w") as fp:
        json.dump(tablet_JSON, fp, indent=2, ensure_ascii = True, allow_nan = True)
    print(f"{file_name} created")
    
````
