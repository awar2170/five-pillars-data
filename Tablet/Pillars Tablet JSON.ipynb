## Put it All Together!

````python
# Tabs of the Excel Sheet
tab_names = ["ARTIFICER", "ENCHANTER", "ORACLE"]
null = None

for x in np.arange(0, len(tab_names),1):
    df = pd.read_excel('--DEV-- Tablet Library.xlsx', sheet_name=f'{tab_names[x]}', skiprows = 1)
    tab = tab_names[x]
    hidden_boolean = False
    tablet_JSON = []
    
    # Collect Row Information 
    for y in np.arange(0, len(df),1):
        item_name = str(df["Name"][y])
        computation_name = str(df["Computational Name"][y])
        description_text = str(df["Description"][y])
        unique_str = str(df["Uniqueness"][y])
        computation_limits = str(df["Computational Limits"][y])
        limitations = str(df["Limitations"][y])
        grade = int(df["Grade"][y])
                
        if tab == "ENCHANTER" or tab == "ORACLE":
            destroy_boolean = True
        else: 
            destroy_boolean = False

        
        # Curse
        if "Curse" in item_name:
            curse_boolean = True
        else:
            curse_boolean = False
        
        
        # Capcity Effect
        if "Capacity Effect" in df:
            capacity = int(df["Capacity Effect"][y])*-1
        else: 
            capacity = None
    
    # Create the JSON Object 
        One_JSON_Object = {"property_name": item_name, "property_computation_name": computation_name, 
                           "property_grade": grade, "property_size": capacity, "property_source": tab, 
                           "property_ability": null, "property_monster": null, "property_type": null, 
                           "property_computation": null, "property_value": null, 
                           "property_limitations": limitations, "property_computation_limits": computation_limits, 
                           "property_description": description_text, "property_unique": unique_str,
                           "property_curse": curse_boolean, "property_destroyable": destroy_boolean, 
                           "property_hidden": hidden_boolean}

        tablet_JSON.append(One_JSON_Object)
        
    file_name = tab + "_tablet.json"
    
    # print(f"{file_name} successfully created")
    # print(tablet_JSON)

    with open(file_name, "w") as fp:
        json.dump(tablet_JSON, fp, indent=2, ensure_ascii = True, allow_nan = True)
    print(f"{file_name} created")
    
````
